Посмотрим с разных сторон на равномерность групп, вначале -  как по группам шли покупки по количеству и по суммам.

```python
location = ['A', 'B']
i = 0
fig, ax = plt.subplots()
for plat in location:
    orders_data_group = orders_data.query('group == @plat')
    orders_data_group = orders_data_group.groupby('date').agg(count_revenue=('revenue','count'))
    orders_data_group.plot(y='count_revenue', style='o-', ylim=(0, 40), grid=True, figsize=(12, 6), label = location[i], ax=ax)
    i += 1
plt.xlabel('дни')
plt.ylabel('количество покупок в сутки')
plt.title('количество покупок в группах по дням')
plt.show()
#https://pyprog.pro/mpl/mpl_adding_a_legend.html
```

Количество покупок не всегда идет похоже, например с 06.08.2019 по 11.08.2019. И это некий не совсем понятный период.

Переменная sampleA собираем из двух частей. Первая - список тех, кто сделал транзакции и список тех, кто не делал транзации, там количество транзакций записываем нулевым. <br/>
Причем количество нулей равно разнице между суммой посетителей и количеством записей о заказах.

```python
sampleA = pd.concat([orders_clear_group_a['transaction_count'],pd.Series(
    0, index=np.arange(visitors_clear_group_a['visitors_day'].sum() - len(orders_clear_group_a['transaction_count'])), name='transaction_count')],axis=0)
#display(sampleA)
```

И тоже самое делаем по группе В:

```python
orders_clear_group_b = orders_data_clear[orders_data_clear['group']=='B'].groupby('visitorid', as_index=False).agg({'transactionid' : pd.Series.nunique})
orders_clear_group_b.columns = ['visitorid', 'transaction_count']

visitors_clear_group_b = visitors_data_clear[visitors_data_clear['group']=='B'][['date', 'visitors_clear']]
visitors_clear_group_b.columns = ['date', 'visitors_day']
```

```python
sampleB = pd.concat([orders_clear_group_b['transaction_count'],pd.Series(
    0, index=np.arange(visitors_clear_group_b['visitors_day'].sum() - len(orders_clear_group_b['transaction_count'])), name='transaction_count')],axis=0)
#display(sampleB)
```

Считаем средние конверсии по группам.

```python
conversion_a_mean_1 = sampleA.sum() / sampleA.count()
print('Средняя конверсия по группе A: {:.2%}'.format(conversion_a_mean_1))
conversion_b_mean_1 = sampleB.sum() / sampleB.count()
print('Средняя конверсия по группе B: {:.2%}'.format(conversion_b_mean_1))
```

Принимаем гипотезы: <br/>
нулевая гипотеза о том, что статистически значимых различий в конверсии между группами нет;<br/>
вторая гипотеза о том, что статистически значимые различия в конверсии между группами есть

```python
alpha=0.05
p_value = st.mannwhitneyu(sampleA, sampleB)[1]
print('p-значение: ', p_value)
if (p_value < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Стало быть по U-критерию Манна — Уитни (англ. Mann–Whitney U-test) статистически значимые различия в конверсии между группами есть.

```python
alpha=0.05
p_value = st.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')[1]
print('p-значение: ', p_value)
if (p_value < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Но есть и иные факты:<br/>
Зафиксированы посетители, одновременно попавшие в обе группы. Да, это всего 3,7% от количества всех покупок и 4,3% от общей суммы покупок. Но это изначально некорректная ситуация для проведения теста и мы достоверно не знаем в какой пропорции смотрели разные версии не купившие посетители и пока точного ответа на этот вопрос нет;<br/>
График кумулятивной выручки, как и остальные графики показывают, что практически все превосходство группы В возникло в период с 06.08.2019 по 11.08.2019 и у нас нет данных, что происходило в тот период.<br/><br/>
Поэтому считаю, что проведение данного теста надо считать некорректным из-за нестрогого разделения аудиторий. Эту проблему необходимо вначале устранить.<br/>
Необходимо также провести дополнительное исследование событий с 06.08.2019 по 11.08.2019.<br/><br/>
