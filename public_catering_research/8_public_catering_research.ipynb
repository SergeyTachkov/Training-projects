```python
def take_pic(data, legend_labels, title, y, x, ylabel='%'):
    fig, ax = plt.subplots(1, 2, figsize=(16, 10))

    silent_labels = [' '] * len(data.values)
    
    
    cmap = sns.color_palette('Blues_r', len(data.values))
    
    data.plot.pie(shadow=True, autopct='%1.1f%%', \
                  labels=silent_labels, ax=ax[0], colors=cmap, subplots=True)
    ax[0].legend(loc=2, labels=legend_labels)
    ax[0].set_title(title, fontsize=12)
    ax[0].set_ylabel(ylabel)
    
    sns.barplot(y=y, x=x, data=data, estimator=sum, orient='h', ax=ax[1], palette=cmap)
    ax[1].set_ylabel(' ')
    ax[1].set_xlabel(' ')
    
    plt.show()
```

```python
rest_data_0_type_graph = (rest_data_0
                 .pivot_table(index='object_type', values='id', aggfunc='count')
                 .sort_values('id', ascending=False))

take_pic(data=rest_data_0_type_graph, legend_labels=rest_data_0_type_graph.index, \
         title='Объекты с 0-количеством мест по типу', y=rest_data_0_type_graph.index, x='id')
```

```python
def take_pic(data, legend_labels, title, y, x, ylabel='%'):
    fig, ax = plt.subplots(1, 2, figsize=(16, 10))

    silent_labels = [' '] * len(data.values)
    
    
    cmap = sns.color_palette('Blues_r', len(data.values))
    
    data.plot.pie(shadow=True, autopct='%1.1f%%', \
                  labels=silent_labels, ax=ax[0], colors=cmap, subplots=True)
    ax[0].legend(loc=2, labels=legend_labels)
    ax[0].set_title(title, fontsize=12)
    ax[0].set_ylabel(ylabel)
    
    sns.barplot(y=y, x=x, data=data, estimator=sum, orient='h', ax=ax[1], palette=cmap)
    ax[1].set_ylabel(' ')
    ax[1].set_xlabel(' ')
    
    plt.show()
```

Сделаем еще по U-критерию Манна — Уитни (англ. Mann–Whitney U-test), так как распределения отличаются от нормального и не зависят между собой.

Принимаем гипотезы: <br/>
нулевая гипотеза о том, что статистически значимых различий в распределении посадочных мест между группами нет; <br/>
вторая гипотеза о том, что статистически значимые различия в распределении посадочных мест между группами есть


```python
rest_data_top10_values = rest_data_sss_top10_no0['number'].values
rest_data_values = rest_data_sss_no0['number'].values

alpha=0.01 #https://habr.com/ru/company/skillfactory/blog/510688/
p_value = st.mannwhitneyu(rest_data_top10_values, rest_data_values)[1]

print('p-значение: ', p_value)
if (p_value < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Получается, что достаточно пограничное условие у нас. <br/>
На alpha=0.05 - на таком уровне статистической значимости результат был бы иным. То есть, распределения достаточно близкие.

Считаем, что статистически значимых различий в распределении посадочных мест между группами нет.
