```python
ecommerce_data_alert = ecommerce_data.groupby(
    'customer_id').agg(order_count=('order_id','count'), product_total=('total','sum'), order_date_first=(
    'date_0','first'), order_date_last=('date_0','last')).reset_index() # делаем дополнительный индекс в слайсе
ecommerce_data_alert = ecommerce_data_alert.sort_values('product_total', ascending = False)
ecommerce_data_alert_date = ecommerce_data_alert.copy() # уйдем от слайса снова к датафрейму

ecommerce_data_alert_date['order_mean'] = ecommerce_data_alert_date['product_total'
                                                                   ] / ecommerce_data_alert_date['order_count']

ecommerce_data_alert_date['period'] = (ecommerce_data_alert_date['order_date_last'] - ecommerce_data_alert_date[
    'order_date_first']) / (ecommerce_data_alert_date['order_count'] - 1)

ecommerce_data_alert_date['warning'] = ecommerce_data_alert_date['order_date_last'
                                                                ] + ecommerce_data_alert_date['period']

display(ecommerce_data_alert_date.head(5))
```

```python
def take_pic_4(data, legend_labels, title, y, x, ylabel='доля'):
    fig, ax = plt.subplots(1, 2, figsize=(14, 7))

    silent_labels = [' '] * len(data.values)
    
    
    cmap = sns.color_palette('Blues_r', len(data.values))
    
    data.plot.pie(shadow=True, autopct='%1.1f%%', \
                  labels=silent_labels, ax=ax[0], colors=cmap, subplots=True)
    ax[0].legend(loc=2, labels=legend_labels)
    ax[0].set_title(title, fontsize=12)
    ax[0].set_ylabel(ylabel)
    
    sns.barplot(y=y, x=x, data=data, estimator=sum, orient='h', ax=ax[1], palette=cmap)
    ax[1].set_ylabel(' ')
    ax[1].set_xlabel(' ')
    for i, val in enumerate(data.values):
        color = '#3E5F8A'
        ax[1].text(val +0.005, i +0.09, float("%0.2f"%val), color=color, fontdict={'fontweight':900, 'size':10})
    
    plt.show()
```

Пишем функция для определения визита в будние дни или выходные. 

```python
def weekend(weekday):
    if weekday == 5:
        return 1
    if weekday == 6:
        return 1
    else:
        return 0
```

Добавляем столбец weekend, где 1 - выходные, 0 - будни.

```python
ecommerce_data['weekend'] = ecommerce_data['weekday'].apply(weekend)

ecommerce_data = ecommerce_data.sort_values(by='weekday', ascending=False)
display(ecommerce_data.head(5))
```

Давайте аналогично сделаем с рабочими часами и вне данного периода.

```python
def work_hour(hour):
    if hour >= 0 and hour <= 9:
        return 0
    if hour >= 18 and hour <= 23:
        return 0
    else:
        return 1
```

```python
ecommerce_data['w_hour'] = ecommerce_data['hour'].apply(work_hour)

ecommerce_data = ecommerce_data.sort_values(by='w_hour', ascending=False)
display(ecommerce_data.sample(5))
```

Графиков предстоит построить много, напишем снова функцию. 

```python
def take_pic_dist_box_3(data, columns_name, title, x_label):
    fig, ax = plt.subplots(figsize = (16, 4))
    
    # список кластеров:
    cluster_list = data['cluster_km'].unique()  # берем уникальные значения
    cluster_list = np.sort(cluster_list) # сортируем
    
    cluster_data = [data.loc[data['cluster_km'] == i] for i in cluster_list] # каждому кластеру свой датафрейм
    
    for cluster in cluster_data:
        sns.distplot(cluster[columns_name], label = cluster['cluster_km'].unique(),
                     hist=False, ax = plt.subplot(1, 2, 1))
        plt.xlabel(x_label)
        plt.ylabel('распределение')
        plt.legend()

    sns.boxplot(y = data[columns_name], x = data['cluster_km'], ax = plt.subplot(1, 2, 2)) 
    plt.xlabel('')
    plt.ylabel(x_label)

    plt.show()
```

<b>3.5. Проверка гипотезы о статистическом различии общей суммы продаж наших кластеров.</b> <br/>
</div>


Ну и давайте проверим уровень статистической значимости нашего результата. Будем использовать U-критерий Манна — Уитни (англ. Mann–Whitney U test) — статистический критерий, используемый для оценки различий между двумя независимыми выборками по уровню какого-либо признака, измеренного количественно. 

Принимаем гипотезы:<br/>
нулевая гипотеза о том, что статистически значимых различий в совокупностях по общей сумме покупок между кластерами нет;<br/>
вторая гипотеза о том, что статистически значимые различия в совокупностях по общей сумме покупок между кластерами есть

```python
alpha = 0.05 
results = stats.mannwhitneyu(ecommerce_data_cluster_km_no2[ecommerce_data_cluster_km_no2['cluster_km'] == '0']['total'],
                             ecommerce_data_cluster_km_no2[ecommerce_data_cluster_km_no2['cluster_km'] == '1']['total'])[1]

print('p-value:', "{0:.2f}".format(results))
if (results < alpha):
    print('Отвергаем нулевую гипотезу. Общая сумма покупок 0-го кластера отличается от суммы 1-го кластера')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Общая сумма 0-го кластера не отличается от суммы 1-го кластера')
```

```python
ecommerce_data.info()
ecommerce_data_order_count_day = ecommerce_data.groupby('date_2').agg(
    order_id_count=('order_id','count'), customer_id_count=('customer_id','count'))
ecommerce_data_order_count_day['delta'] = ecommerce_data_order_count_day[
    'order_id_count'] - ecommerce_data_order_count_day['customer_id_count']
ecommerce_data_order_count_day = ecommerce_data_order_count_day.sort_values('delta', ascending = False)
display(ecommerce_data_order_count_day.head(5))
```
