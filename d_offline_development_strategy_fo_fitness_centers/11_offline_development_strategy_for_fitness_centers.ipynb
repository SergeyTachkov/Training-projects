```python
def take_pic_dist_box_1(data, columns_name, title, x_label):
    fig, ax = plt.subplots(figsize = (16, 4))
    sns.distplot(data[data['churn']==1][columns_name], label = 'отток', color = 'darkorange', ax = plt.subplot(1, 2, 1))
    sns.distplot(data[data['churn']==0][columns_name], label = 'актив', color = 'darkblue', ax = plt.subplot(1, 2, 1)) 
    
    plt.xlabel(x_label)
    plt.ylabel('распределение')
    plt.legend()

    sns.boxplot(y = columns_name, x = 'churn', data = data, ax = plt.subplot(1, 2, 2)) 
    plt.xlabel('')
    plt.ylabel(x_label)
    plt.xticks(np.arange(2), ('Актив', 'Отток'))
    
   

    fig.suptitle(title)
    plt.show()
```

Поэтому сделаем вначале на общем датасете без изменений, а потом посмотрим с стандартизацией и удалением выбросов и уже потом посмотрим без быстроуходящих клиентов.


Сделаем функцию для вывода метрик:

```python
# функция для вывода метрик
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('- Accuracy: {:.3f}'.format(accuracy_score(y_test, y_pred)))
    print('- Precision: {:.3f}'.format(precision_score (y_test, y_pred)))
    print('- Recall: {:.3f}'.format(recall_score(y_test, y_pred)))
```

```python
def take_pic_dist_box_3(data, columns_name, title, x_label):
    fig, ax = plt.subplots(figsize = (16, 4))
    
    # список кластеров:
    cluster_list = data['cluster_km'].unique()  # берем уникальные значения
    cluster_list = np.sort(cluster_list) # сортируем
    
    cluster_data = [data.loc[data['cluster_km'] == i] for i in cluster_list] # каждому кластеру свой датафрейм
    
    for cluster in cluster_data:
        sns.distplot(cluster[columns_name], label = cluster['cluster_km'].unique(),
                     hist=False, ax = plt.subplot(1, 2, 1))
        plt.xlabel(x_label)
        plt.ylabel('распределение')
        plt.legend()

    sns.boxplot(y = data[columns_name], x = data['cluster_km'], ax = plt.subplot(1, 2, 2)) 
    plt.xlabel('')
    plt.ylabel(x_label)

    plt.show()
    
```

<b>6.1. Новички, исходя из данных этого отчета выше - катастрофически плохо адаптируются к сервису.</b> <br/>Считаю, что сервис должен написать целую программу адаптации, которую должен использовать его тренерский состав. В основном, это дополнительная мотивация и одобрение всех мало-мальских действий новичка. Можно даже выдавать новичку при входе в сервис первые два месяца какой-то яркий атрибут, типа яркого бесплатного полотенца - все тренера будут таких сразу лучше выделять в зале занятий и отдельно усиленно наблюдать, подсказывать (без доп.оплаты), хвалить и подбадривать. И даже говорить, что обязательно будут ждать его на следующей тренировке! ))<br/><br/>
<b>6.2. Лонгируемая подписка вместо абонемента на 1 месяц.</b><br/>
Советую не продавать разовые абонементы на 1 месяц, а делать подписку на 1 месяц с оплатой по карте и с автоматической пролонгацией и возможностью отказа. Да еще и с возможностью приостановки действия. Думаю, такая схема будет более рентабельной относительно существующей.<br/><br/>
<b>6.3. Вести учет каждого посещения клиента.</b><br/>
Большинство клиентов ходит на занятия по определенным дням, выбранным им самими. Средствами подобной аналитики можно предсказать дату следующего посещения клиента. И ставить алерты с учетом стандартных сезонов отпусков с целью дополнительного мотивирущего контакта со стороны сервиса с таким клиентом. Это может быть предложение скидки, новость о новом появившемся сервисе или новой возможности. Просто сказать, что мы вас ждем ))<br/><br/>
<b>6.4. Проведение рекламных кампаний именно с учетом geo потенциального клиента.</b><br/>
Надо бы вести рекламные кампании четко по аудитории тех, кто живет или появляется в районе сети определенное количество раз за месяц. Тем более, что это вполне возможно в некоторых рекламных системах.<br/><br/>
<b>6.5. Обойти все крупные предприятия района с предложением.</b><br/>
Постараться поговорить со всеми предприятиями, привлекая обязательно к разговору и главных руководителей предприятий на предмет партнерской программы.
